Llama 2

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_path ="meta-llama/Llama-3.2-3B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_path)
```

```python
#8-bit quantization
import torch
from transformers import BitsAndBytesConfig

# model = AutoModelForCausalLM.from_pretrained(
#     model_path, load_in_8bit=True, device_map="auto")

# 4-bit quant
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

model = AutoModelForCausalLM.from_pretrained(
    model_path,
    quantization_config=bnb_config,
    device_map="auto"
)
```

```python
import torch
import transformers

from transformers import pipeline
gen_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=256,
    #dtype="float16",
    temperature=0.3,
    repetition_penalty=1.1,
    return_full_text=False,
    do_sample=True
    # Assuming GPU usage, change to -1 if using CPU
)
```

```python
from langchain_community.llms import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=gen_pipeline)
```

```python
from langchain_core.prompts import ChatPromptTemplate

mcq_prompt_template = """<|begin_of_text|><|start_header_id|>system<|end_header_id|>
You are an expert educator. Generate high-quality MCQs where distractors are plausible but clearly incorrect based ONLY on the provided context. Avoid "All of the above" unless absolutely necessary.<|eot_id|>

<|start_header_id|>user<|end_header_id|>
Context: {context}

Generate ONE MCQ. Follow this format:
Question: [Text]
A) [Option]
B) [Option]
C) [Option]
D) [Option]
Correct Answer: [Letter]
Explanation: [Concise reason why the answer is correct based on the text]

MCQ:<|eot_id|>
<|start_header_id|>assistant<|end_header_id|>
Question:"""
```

```python
mcq_prompt = ChatPromptTemplate.from_template(mcq_prompt_template)
```

```python
from langchain_core.output_parsers import StrOutputParser

mcq_chain= mcq_prompt | llm | StrOutputParser()

def generate_mcqs(topic, num_questions=5):

  docs=retriver.invoke(topic)
  generated_mcqs = []


  for i in range(min(num_questions, len(docs))):
    context_chunk = docs[i].page_content

    response = mcq_chain.invoke({"context": context_chunk})
    generated_mcqs.append(response)

  return generated_mcqs
```

```python
mcqs = generate_mcqs("Fibrocystic breast disease", num_questions=5)

for i, mcq in enumerate(mcqs, start=1):
    print(f"\nMCQ {i}:\n{mcq}")
```
